```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df['Order Year'] = df['Order Date'].dt.year
```

```python
df = df[['Order ID', 'Order Date', 'Ship Date', 'Order Year', 'Order Month', 'Order DayName',
         'Customer ID', 'Customer Name', 'Segment', 'Country', 'City', 'State', 'Region',
         'Product ID', 'Category', 'Sub-Category', 'Product Name',
         'Sales', 'Quantity', 'Discount', 'Profit']]

```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending = False).head(10).reset_index()
top_products
```

```python
top_products.plot(kind = 'barh', figsize = (10,6),title = 'Top 10 products by sales')
plt.xlabel('total_sales')
plt.ylabel('Product Name')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
fig = px.bar(
    top_products,
    x='Sales',
    y='Product Name',
    color='Sales',
    title='Top 10 products by sales'
)

fig.update_layout(
    xaxis_title='total_sales',
    yaxis_title='Product Name',
    xaxis_tickangle=-45,
    height=500
)

fig.show()
```

```python
fig.write_html('Top 10 products by sales.html')
```

```python
monthly_sales = df.groupby(['Order Year','Order Month'])['Sales'].sum().reset_index()
monthly_sales['year-month'] = (monthly_sales['Order Year'].astype(str) + '-' + monthly_sales['Order Month'].astype(str).str.zfill(2))
monthly_sales['year-month'] = pd.to_datetime(monthly_sales['year-month'])
monthly_sales.sort_values('year-month', inplace=True)
monthly_sales['year-month'] = monthly_sales['year-month'].dt.to_period('M').astype(str)
```

```python
monthly_sales
```

```python
print(monthly_sales.dtypes)
```

```python
plt.figure(figsize=(20, 10))
plt.bar(monthly_sales['year-month'], monthly_sales['Sales'], color='skyblue')

plt.xticks(rotation=90)
plt.title('Monthly Sales Trend')
plt.xlabel('Year-Month')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

```python
monthly_sales_profit = df.groupby(['Order Year','Order Month'])['Profit'].sum().reset_index()
monthly_sales_profit['year-month'] = (monthly_sales_profit['Order Year'].astype(str) + '-' + monthly_sales_profit['Order Month'].astype(str).str.zfill(2))
monthly_sales_profit['year-month'] = pd.to_datetime(monthly_sales_profit['year-month'])
monthly_sales_profit.sort_values('year-month', inplace=True)
monthly_sales_profit['year-month'] = monthly_sales_profit['year-month'].dt.to_period('M').astype(str)
df['year-month'] = monthly_sales_profit['year-month']
monthly_sales_profit
```

```python
colors = monthly_sales_profit['Profit'].apply(lambda x: 'green' if x >= 0 else 'red')
plt.figure(figsize=(20, 10))
plt.bar(monthly_sales_profit['year-month'], monthly_sales_profit['Profit'], color=colors)
plt.xticks(rotation=90)
plt.title('Monthly profits Trend')
plt.xlabel('Year-Month')
plt.ylabel('Total Profits')
plt.tight_layout()
plt.show()
```

```python
fig = px.bar(
    monthly_sales_profit,
    x='year-month',
    y='Profit',
    color='Profit',
    title='Monthly profits Trend'
)

fig.update_layout(
    xaxis_title='Year-Month',
    yaxis_title='Total Profits',
    xaxis_tickangle=-45,
    height=500
)

fig.show()
```

```python
fig.write_html('Monthly profits Trend.html')
```

```python
category_profit = df.groupby(['Category', 'Sub-Category'])['Profit'].sum().sort_values(ascending=False).reset_index().head(10)
category_profit['Label'] =  category_profit['Category'] + '-' + category_profit['Sub-Category']
category_profit['Label']
```

```python
fig = px.bar(
    category_profit,
    x='Profit',
    y='Label',
    orientation='h',
    color='Label',
    color_continuous_scale='Reds', 
    title='Top 10 Most Profitable Product Sub-Categories'
)

fig.update_layout(
    xaxis_title='Total Profit',
    yaxis_title='Category - Sub-Category',
    yaxis={'categoryorder': 'total ascending'},
    height=600
)

fig.show()
```

```python
fig.write_html('Top 10 Most Profitable Product Sub-Categories.html')
```

```python
plt.figure(figsize=(12, 6))
plt.barh(category_profit['Label'], category_profit['Profit'], color='forestgreen')
plt.title('Top 10 Most Profitable Product Sub-Categories')
plt.xlabel('Total Profit')
plt.ylabel('Category - Sub-Category')
plt.gca().invert_yaxis()  
plt.tight_layout()
plt.show()
```

```python
less_category_profit = df.groupby(['Category', 'Sub-Category'])['Profit'].sum().sort_values(ascending=False).reset_index().tail(10)
less_category_profit['Label'] =  less_category_profit['Category'] + '-' + less_category_profit['Sub-Category']
less_category_profit['Label']
```

```python
colors = less_category_profit['Profit'].apply(lambda x: 'green' if x >= 0 else 'red')

plt.figure(figsize=(12, 6))
plt.barh(less_category_profit['Label'], less_category_profit['Profit'], color=colors)
plt.title('Top 10 Most Profitable Product Sub-Categories')
plt.xlabel('Total Profit')
plt.ylabel('Category - Sub-Category')
plt.gca().invert_yaxis()  
plt.tight_layout()
plt.show()
```
